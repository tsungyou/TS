```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

url = "https://histock.tw/stock/rank.aspx?p=all"
res = requests.get(url)
soup = BeautifulSoup(res.text, "lxml")
altrow = soup.find_all("tr", {"class": "alt-row"})
rows = []
for index, i in enumerate(altrow):
    subrow = []
    for ind, k in enumerate(i):
        content = k.text.replace("\n", "").replace("%", "").replace(",", "")
        if ind == 0 or ind == 14:
            pass
        if ind >= 6 and ind < 14:
            content = float(content)
        subrow.append(content)
    rows.append(subrow)
```

```python
import yfinance as yf
```

```python
df = pd.DataFrame(rows)
df.columns = ["1Nothing", "code", "codename", "value-change", "pct-change", "week-pct-change", \
              "high-low", "Open", "High", "Low", "Close", "last-close", "volume-ticket", "volume_price", "Nothing"]

dfs = df.sort_values(by="volume_price", ascending=False).iloc[:, :]
codes = list(dfs['code'])

df_all = pd.DataFrame()
for i in codes:
    market = "TW"
    ticker_market = i + "." + market
    df_price = yf.download(ticker_market, start="2019-01-01", progress=False)
    if len(df_price) == 0:
        market = "TWO"
        ticker_market = i + "." + market
        df_price = yf.download(ticker_market, start="2019-01-01", progress=False)
    df_price['ticker'] = i
    df_price['market'] = market
    df_price['ticker_market'] = ticker_market
    df_all = pd.concat([df_all, df_price])
df_all.to_parquet("price_top290.parquet")
```
