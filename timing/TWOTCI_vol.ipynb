```python
import pandas as pd
import os
import matplotlib.pyplot as plt
def get_cn_price_df():
    list_ = os.listdir(cn_price)
    list_ = [os.path.join(cn_price, file) for file in list_]
    list_parquet = [pd.read_parquet(file) for file in list_]
    df = pd.concat(list_parquet)
    return df

cn_price = "../db/cn/"
df = get_cn_price_df()
# df1 = pd.read_parquet("../db/cn/cn_price_2024.parquet")
twse = df[df['code'] == "TWSE Index"][['da', 'cl', 'op', 'vol']].set_index("da").sort_index(ascending=True)
twse
```
