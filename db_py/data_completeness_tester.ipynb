```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
directories = ['tw/pb_ratio', "tw/price"]
parquet_files = [os.path.join(directories[1], f) for f in os.listdir(directories[1]) if f.endswith(".parquet")] 
dfs = []
parquet_files.remove("tw/price/2021.parquet")
parquet_files.remove("tw/price/2019.parquet")
# parquet_files.remove("tw/price/2018.parquet")
print(parquet_files)
for file in parquet_files:
    df = pd.read_parquet(file)
    dfs.append(df)
df_concat = pd.concat(dfs)
pivoted = df_concat.pivot(index='da', values="cl", columns="ticker")

```

```python
values_dict = {
    "cl": "close",
    "op": "open",
    "vol(turnover)": "volume",
    'vol(volume)': "volume(share)",
    "yield": "yield",
    "pe_ratio": "pe_ratio",
    "pb_ratio": "pb_ratio",
}
values = 'vol(turnover)'
pivoted = df_concat.pivot(index='da', values=values, columns="ticker")
pivoted.replace("--", None, inplace=True)
pivoted = pivoted.ffill()
pivoted = pivoted.astype(float)
pivoted.astype(float).to_parquet(f"tw/pdata/{values_dict[values]}.parquet")
pivoted_pct = pivoted.pct_change().dropna(how="all")
pivoted_pct.to_parquet(f'tw/pdata/{values_dict[values]}_pct.parquet')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 目標網址 (公開資訊觀測站，台積電的財務報表)
url = "https://mops.twse.com.tw/mops/web/t05st31"

# 發送HTTP GET請求
params = {
    'co_id': '2330',
    'year': '2022',
    'season': '01'
}
response = requests.post(url, data=params)
response.encoding = 'utf-8'

# 解析HTML
soup = BeautifulSoup(response.text, 'html.parser')

# 找到財務報表所在的表格
table = soup.find('table', {'class': 'hasBorder'})

# 解析表格內容
data = []
headers = [th.text for th in table.find_all('th')]
rows = table.find_all('tr')[1:]  # 跳過標題列
for row in rows:
    cols = [td.text.strip() for td in row.find_all('td')]
    data.append(cols)

# 將資料轉換成Pandas DataFrame
df = pd.DataFrame(data, columns=headers)

# 顯示資料
print(df)
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# 設定 WebDriver 路徑
driver_path = '/path/to/chromedriver'  # 替換為 ChromeDriver 的實際路徑

# 初始化 WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 啟動無頭模式
driver = webdriver.Chrome(executable_path=driver_path, options=options)

# 目標網址
url = "https://mops.twse.com.tw/mops/web/t05st31"

# 發送HTTP GET請求
driver.get(url)

# 填寫表單
co_id_input = driver.find_element(By.NAME, 'co_id')
year_input = driver.find_element(By.NAME, 'year')
season_input = driver.find_element(By.NAME, 'season')

co_id_input.send_keys('2330')
year_input.send_keys('2022')
season_input.send_keys('01')

# 提交表單
submit_button = driver.find_element(By.XPATH, '//input[@type="button" and @value="查詢"]')
submit_button.click()

# 等待網頁加載
time.sleep(5)

# 獲取網頁內容
html_content = driver.page_source

# 解析HTML
soup = BeautifulSoup(html_content, 'html.parser')

# 找到財務報表所在的表格
table = soup.find('table', {'class': 'hasBorder'})

# 檢查表格是否存在
if table:
    # 解析表格標題
    headers = [th.text.strip() for th in table.find_all('th')]

    # 解析表格數據
    data = []
    rows = table.find_all('tr')[1:]  # 跳過標題列
    for row in rows:
        cols = [td.text.strip() for td in row.find_all('td')]
        if cols:
            data.append(cols)

    # 將資料轉換成Pandas DataFrame
    df = pd.DataFrame(data, columns=headers)

    # 顯示資料
    print(df)
else:
    print("未找到財務報表")

# 關閉 WebDriver
driver.quit()
```

```python
import requests
import zipfile
import os

# 下载 ChromeDriver
url = "https://storage.googleapis.com/chrome-for-testing-public/126.0.6478.126/mac-x64/chromedriver-mac-x64.zip"
response = requests.get(url)

# 将文件保存到本地
with open("chromedriver-mac-x64.zip", "wb") as file:
    file.write(response.content)

# 解压缩文件
with zipfile.ZipFile("chromedriver-mac-x64.zip", "r") as zip_ref:
    zip_ref.extractall("chromedriver_mac")

# 获取 ChromeDriver 路径
chromedriver_path = os.path.abspath("chromedriver_mac/chromedriver-mac-x64/chromedriver")
print(f"ChromeDriver 路径: {chromedriver_path}")
```
